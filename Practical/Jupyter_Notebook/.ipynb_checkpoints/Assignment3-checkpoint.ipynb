```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nbaallelo.csv')
```

```python
df = df.drop("notes",axis=1)
```

```python
df.shape
```

```python
df
```

```python
# It is a countplot representing number of NBA matches played in this every year
# The sudden increase in number of matches played in year 1967 and 1968 is dur to expansion of league from 10 teams to 12 teams.
# Also for each played match there are two records in dataset, one for winning team and one for losing team.
```

```python
plt.figure(figsize=(20,14))
sns.countplot(data=df, x="year_id")
plt.xticks(rotation=45)
plt.show()
```

```python
df[df["date_game"] == '11/1/1946']
```

```python
winningGame = df[df["game_result"]=="W"]
```

```python
winningGame
```

```python
# It is a barplot representing number of matches won by each team. The team with highest winnings is BOS and the team with less winnings is SDS 
```

```python
id_counts = winningGame['team_id'].value_counts().reset_index()

id_counts.columns = ['team_id', 'count']

sorted_df = id_counts.sort_values(by='count', ascending=False)

plt.figure(figsize=(20,30))
sns.barplot(data=sorted_df, x='count', y='team_id')
plt.show()
```

```python
df[df['forecast'] > 0.5]
```

```python
# It is a piechart represnting the accuracy of forcast. It represesnts number actual wins when forecast was win and number of actual loss even forecast was loss
```

```python
plt.figure(figsize=(20,14))
count = df[df['forecast'] > 0.5]['game_result'].value_counts()
plt.pie(count,labels = count.index,autopct='%.1f%%')
plt.show()
```

```python
df
```

```python
# It is a scatterplot reprenting number of points scored by each team in different matches
# A general range of goals scored in match is 50 to 150
```

```python
plt.figure(figsize=(20,20))
sns.scatterplot(data=df,x='pts',y='team_id',hue='game_result')
plt.show()
```

```python
# It is a heatmap represnting how the features in dataset are related to each other
```

```python
plt.figure(figsize=(20, 14))
sns.heatmap(df.corr(numeric_only=True))
plt.plot()
```

```python
# It is a line plot representing the fluctuations of elo rating of team SDS over time
```

```python
plt.figure(figsize=(20, 14))
sns.lineplot(data=df[df['team_id']=='SDS'], x="elo_i", y="elo_n")
plt.plot()
```

```python
#It is a histplot representing the number of matches won and lose by team BOS in home ground and away ground
# team has won more matches in home grounds
```

```python
sns.histplot(data=df[df['team_id']=='BOS'], x="game_location", hue="game_result")
```

```python
# it is joint plot represnting realtion between game results, points score and match results
# Mostly if team has scored above 100 points, they have won game
# and forcast has not much effect on game results
```

```python
plt.figure(figsize=(20, 14))
sns.jointplot(data=df, x="pts", y="forecast", hue="game_result")
plt.plot()
```

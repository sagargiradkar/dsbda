```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
```

## Univariant

```python
x=house[["area"]]
y=house[["price"]]
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.75)
```

```python
scaler = MinMaxScaler()
```

```python
x_train
```

```python
x_train= scaler.fit_transform(x_train)
```

```python
x_train
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_test
```

```python
x_test= scaler.fit_transform(x_test)
```

```python
x_test
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(x_train, y_train)
```

```python
y_pred = linear_regressor.predict(x_test)
```

```python
linear_regressor.coef_
```

```python
linear_regressor.intercept_
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
mb.title("Area-Price")
mb.xlabel("area")
mb.ylabel("price")
mb.scatter(x_test,y_test,label='actual',color='r')
mb.plot(x_test,y_pred,label='predicted',color='g')
mb.legend()
```

### Prediction

```python
newxtest=[[7420]]
newxtest= scaler.transform(newxtest)
predicted = linear_regressor.predict(newxtest)
predicted
```

```python
mse=mean_squared_error(y2_test,y_pred1)
mse
```

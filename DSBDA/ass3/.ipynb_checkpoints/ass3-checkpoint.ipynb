```python
sex_counts = df["Sex"].value_counts()
sex_labels = sex_counts.index
plt.pie(sex_counts, labels=sex_labels,  autopct='%.1f%%', startangle=140)
plt.title("Distribution of Sex")
```

```python
df.Sex = df.Sex.map({'female': 0, 'male': 1})
df.Embarked = df.Embarked.map({'S': 0, 'C': 1, 'Q': 2})
```

```python
corr_matrix = df.corr()
```

```python

sns.heatmap(corr_matrix,annot=True, fmt=".1f")
```

```python

# Create the area plot
sns.barplot(x=df["Pclass"], y=df["Survived"], color='skyblue', label='Group 1')
sns.barplot(x=df["SibSp"], y=df["Survived"], color='red', label='Group 2')

# Add labels and title
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Area Plot')
plt.legend()

# Show the plot
plt.show()
```
